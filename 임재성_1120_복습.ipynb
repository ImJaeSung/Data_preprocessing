**파생변수** **만들기**

```python
mpg.assign(cby = 0.55*mpg['cty'] + 0.45*mpg['hwy']) #복합연비라는 새로운 변수 생성
```

```python
mpg.assign(cby = 0.55*mpg['cty'] + 0.45*mpg['hwy'], inplace = True) # 원본 데이터에 적용
```

```python
mpg['cby'] = 0.55*mpg['cty'] + 0.45*mpg['hwy'] # 새로운 변수 만드는 방법 2
```

```python
mpg
```

```python
mpg_new = mpg.sort_values('cby', ascending = False) # 새로운 변수 복합연비에 대해 내림차순으로 정렬한 후 새로운 변수에 저장
mpg_new
```

- lambda를 이용한 데이터 전처리

```python
mpg.assign(cby = lambda x : 0.55 * x['cty'] + 0.45 * x['hwy']) #복합연비 변수 만들기
```

```python
mpg.assign(cby = lambda x : 0.55 * x['cty'] + 0.45 * x['hwy'],
           mean = lambda x : (x['cty'] + x['hwy']) / 2) # 복합연비와 두 고속도로 연비와 도시연비의 평균, 2개의 새로운 변수 만들기
```

- 집단별 요약 (group by)

```python
mpg.groupby('manufacturer').agg(mean_hwy = ('hwy', 'mean'))  # 제조사별 고속도로 연비의 평균
```

```python
mpg.groupby('manufacturer').agg(mean_cty = ('cty', 'mean')) #제조사별 도시연비의 평균
```

```python
mpg = pd.read_csv(file_path) 
```

```python
mpg.groupby('year').agg(mean_displ = ('displ', 'mean')) # 제조년도에 따른 배기량의 평균치
```

```python
mpg.groupby('manufacturer').agg(mean_hwy = ('hwy', 'mean'),
                                mean_cty = ('cty', 'mean'),
                                sum_displ = ('displ', 'sum'),
                                n = ('manufacturer', 'count')) #제조사별 여러 변수의 요약 통계 값을 보고 싶은 경우
```

```python
mpg.groupby('manufacturer').mean() # 모든 변수에 대해 제조사별 평균을 구하는 방법
```

```python
mpg.groupby('manufacturer').sum()
```
