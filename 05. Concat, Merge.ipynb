### 데이터 병합
- concat

```python
import pandas as pd
```

```python
df1 = pd.DataFrame({'A':[1, 2, 3],
                    'B':[4, 5, 6]})
df2 = pd.DataFrame({'A':[7, 8, 9],
                    'B':[10, 11, 12]})
```

```python
display(df1)
display(df2)
```

```python
pd.concat([df1, df2],axis = 0)  # 인덱스가 그대로 붙어서 불연속적임
```

```python
pd.concat([df1, df2],axis = 1)  # 컬럼으로 결합
```

```python
## 인덱스를 재정렬할 경우
pd.concat([df1, df2], axis = 0, ignore_index = True)  # ignore_index 매개변수를 이용하여 기존의 데이터프레임의 index를 무시함
```

### 데이터의 차원이 다른경우

```python
df1 = pd.DataFrame({'A':[1, 2, 3],
                    'B':[4, 5, 6]})
df2 = pd.DataFrame({'A':[7, 8],
                    'B':[10, 11]})
```

```python
display(df1)
display(df2)
```

```python
pd.concat([df1, df2], axis = 1) # concat에서는 결측치가 발생할 수 있음
```

- merge

```python
df1 = pd.DataFrame({'A':[1, 2, 3, 4],
                    'B':[5, 6, 7, 8]})
df2 = pd.DataFrame({'A':[1, 2, 3, 4],
                    'C':[10, 11, 12, 13]})
```

```python
display(df1)
display(df2)
```

#### on 인자

```python
pd.merge(df1, df2, on = 'A')
```

```python
pd.merge(df1, df2, on = 'B') # on 에 입력된 컬럼은 병합할 두 데이터프레임의 컬럼에 공통적으로 가지고 있어야 함
```

##### 두 개 이상의 컬럼을 조인 키로 사용

```python
df1 = pd.DataFrame({'A':[1, 2, 3, 4],
                    'B':[5, 6, 7, 8],
                    'D':[0, 0, 0, 3]})
df2 = pd.DataFrame({'A':[1, 2, 3, 4],
                    'B':[5, 6, 7, 8],
                    'E':[1, 1, 1, 0]})
```

```python
pd.merge(df1, df2, on = ['A', 'B']) # 컬럼 A가 같고 컬럼 B가 같은 행만 병합
```

##### 공통 컬럼의 value가 다른 경우

```python
df1 = pd.DataFrame({'A':[1, 2, 30, 40],
                    'B':[5, 6, 7, 8],
                    'D':[0, 0, 0, 3]})
df2 = pd.DataFrame({'A':[1, 2, 3, 4],
                    'B':[5, 6, 7, 8],
                    'E':[1, 1, 1, 0]})
```

```python
display(df1) 
display(df2)
```

```python
pd.merge(df1, df2, on = ['A', 'B']) # 컬럼의 값도 동일해야지만 가져올 수 있다.
```

```python
df1 = pd.DataFrame({'A':[5, 2, 3, 4],
                    'B':[1, 6, 7, 8],
                    'D':[0, 0, 0, 3]})
df2 = pd.DataFrame({'A':[1, 2, 30, 40],
                    'B':[5, 6, 7, 8],
                    'E':[1, 1, 1, 0]})
```

```python
display(df1)
display(df2)
```

```python
pd.merge(df1, df2, on = ['A', 'B']) # A에서 같은 거 고른 후, B에서도 같은지 확인 후 출력
```

#### left_on, right_on

```python
df1 = pd.DataFrame({'A':[1, 2, 3, 4],
                    'B':[5, 6, 7, 8],
                    'D':[0, 0, 0, 3]})
df2 = pd.DataFrame({'a':[1, 2, 3, 4],
                    'B':[5, 6, 7, 8],
                    'E':[1, 1, 1, 0]})
```

```python
display(df1)
display(df2)
```

```python
pd.merge(df1, df2, left_on = 'A', right_on = 'a')  # B_x와 B_y로 나오는 것은 중복이라는 뜻
                                                   # A와 a는 컬럼명이 다르므로 각각 따로 출력
```

```python
pd.merge(df1, df2, left_on = ['A', 'B'], right_on = ['a', 'B']) # B를 추가해서 중복 제거
```

### left_index, right_index
- 인덱스를 조인키로 사용하는 경우

```python
df1 = pd.DataFrame({'A':[1, 2, 3, 4],
                    'D':[0, 0, 0, 3]})
df2 = pd.DataFrame({'A':[1, 2, 3, 4],
                    'E':[1, 1, 1, 0]})
```

```python
display(df1)
display(df2)
```

```python
df1.set_index('A', inplace = True)
```

```python
pd.merge(df1, df2, left_index = True, right_on = 'A') # 왼쪽 데이터프레임에서는 인덱스를, 오른쪽 데이터프레임에서는 컬럼A를 조인키로 사용
```

```python
pd.merge(df1, df2, left_index = False, right_on = 'A') # 왼쪽 데이터프레임에서 조인키를 지정하지 않음
```

```python
display(df1)
display(df2)
```

```python
pd.merge(df1, df2, left_index = True, right_on = 'E') # 왼쪽 데이터프레임은 조인키로 인덱스를, 오른쪽 데이터프레임에서 조인키로 E를 사용했는데 
                                                      # 왼쪽 데이터프레임의 조인키와 오른쪽 데이터프레임의 조인키 중 같은 것이 1이므로 E가 1인 것만 남음
```

### How

```python
pd.merge(df1, df2, on = 'A', how = 'inner') # 교집합 개념
```

```python
pd.merge(df1, df2, on = 'A') # default는 inner
```

```python
pd.merge(df1, df2, on = 'A', how = 'left') # 왼쪽 데이터프레임을 기준으로 병합하므로 오른쪽 데이터프레임에 없는 값은 결측치가 됨
```

```python
pd.merge(df1, df2, on = 'A', how = 'right') # 오른쪽 데이터프레임을 기준으로 병합하므로 왼쪽 데이터프레임에 없는 값은 결측치가 됨
```

```python
pd.merge(df1, df2, on = 'A', how = 'outer') # 두 데이터프레임에 있는 모든 값을 병합하므로 양쪽 데이터프레임에 없는 값은 모두 결측치가 됨
```
