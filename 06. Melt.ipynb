###pd.melt()

- id_vars: 위치를 그대로 유지하는 열의 이름
- value_vars: 행으로 위치를 변경할 (기존의) 열의 이름, 디폴트는 id_vars에서 설정한 열을 제외한 전체
- var_name: value_var 위치로 변경한 (새론운) 열의 이름
- value_name: var_name 위치로 변경한 열의 (새로운) 데이터 컬럼 이름

```python
import pandas as pd
```

```python
file_url = "https://drive.google.com/file/d/1fjTx8R7BlOnnKBS5ypXYo9-SK2j_-1Td/view?usp=share_link" # 구글드라이브 csv파일 이미지 주소
file_path ='https://drive.google.com/uc?id=' + file_url.split('/')[-2] #'/'로 구분했을 때 url_path의 뒤에서 두번째 값이 공유된 파일의 고유값

print("공유파일 링크변환 경로명 : ", file_path)
pew = pd.read_csv(file_path)
```

```python
pew
```

```python
pd.melt(pew, id_vars = 'religion') # reglion 컬럼만 남겨두고 나머지 컬럼은 variable과 value로 취급
```

```python
pd.melt(pew, id_vars = 'religion', var_name = 'income', value_name = 'count') 
# reglion 컬럼만 남겨두고 나머지 컬럼은 income이라는 컬럼명을 가진 variable과 count라는 컬럼명을 가진 value로 취급
```

```python
pd.melt(pew, id_vars = 'religion', value_vars = ['<$10k', '>150k'], var_name = 'income', value_name = 'count')
# reglion 컬럼만 남겨두고 <$10k과 >150k 컬럼만 income이라는 컬럼명을 가진 variable과 count라는 컬럼명을 가진 value로 취급
```
