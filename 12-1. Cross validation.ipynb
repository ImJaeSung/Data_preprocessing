```python
# boxplot을 통해 이상치 탐색
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.gridspec as gridspec

grid = gridspec.GridSpec(5, 4)
plt.figure(figsize= (7, 12))
plt.subplots_adjust(wspace = 0.3, hspace = 0.3)
features = df_1.columns.drop(['상담전화건수', '전화해지여부'])

lower_1 = []
upper_1 = []

for idx, feature in enumerate(features):
    sns.set_style('white')
    ax = plt.subplot(grid[idx])
    sns.boxplot(df_1[feature])
    plt.rc('font', family = 'NanumBarunGothic') 
    ax.set_title(f'{feature}', fontsize = 10)
    plt.xticks(fontsize = 7)
    plt.yticks(fontsize = 7)

    lower_1.append(ax.lines[2].get_ydata()[1])
    upper_1.append(ax.lines[3].get_ydata()[1])
```

```python
grid = gridspec.GridSpec(5, 4)
plt.figure(figsize= (7, 12))
plt.subplots_adjust(wspace = 0.3, hspace = 0.3)
features = df_0.columns.drop(['상담전화건수', '전화해지여부'])

lower_0 = []
upper_0 = []

for idx, feature in enumerate(features):
    sns.set_style('white')
    ax = plt.subplot(grid[idx])
    sns.boxplot(df_0[feature])
    plt.rc('font', family = 'NanumBarunGothic') 
    ax.set_title(f'{feature}', fontsize = 10)
    plt.xticks(fontsize = 7)
    plt.yticks(fontsize = 7)

    lower_0.append(ax.lines[2].get_ydata()[1])
    upper_0.append(ax.lines[3].get_ydata()[1])
```

```python
features = df_1.columns.drop(['상담전화건수', '전화해지여부'])
for idx, feature in enumerate(features):
  df_1.loc[df_1[feature] > upper_1[idx], feature] = upper_1[idx]
  df_1.loc[df_1[feature] < lower_1[idx], feature] = lower_1[idx]
```

```python
features = df_0.columns.drop(['상담전화건수', '전화해지여부'])
for idx, feature in enumerate(features):
  df_0.loc[df_0[feature] > upper_0[idx], feature] = upper_0[idx]
  df_0.loc[df_0[feature] < lower_0[idx], feature] = lower_0[idx]
```

```python
train = pd.concat([df_1, df_0], axis = 0).drop(['가입일', '음성사서함이용'], axis = 1).sort_index()
```

```python
train_input = train.drop(['전화해지여부'], axis = 1)
train_target = train['전화해지여부']
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler()
ss.fit(train_input)
train_scaled = ss.transform(train_input)
```

```python
train_scaled = pd.DataFrame(data = train_scaled, index = train_input.index, columns = train_input.columns)
train_scaled
```

```python
print('feature 들의 평균 값')
print(train_scaled.mean())
print('\nfeature들의 분산 값')
print(train_scaled.var()) 
```

```python
train = pd.concat([train_scaled, train_target], axis = 1)
train
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score, cross_validate
```

```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/BDAA 학회/데이터셋/call_train_임재성87.csv')
```

```python
train_input = df[['log_가입일', 'log_음성사서함이용', '상담전화건수', '주간통화시간', '저녁통화시간', '밤통화시간','주간통화횟수', '저녁통화횟수', '밤통화횟수', '주간통화요금', '저녁통화요금', '밤통화요금']]
train_target = df['전화해지여부']
```

```python
model = [DecisionTreeClassifier(), RandomForestClassifier(), LogisticRegression()]
name = ['DT','RF','LR']

for model, name in zip(model, name):
    print('########## 사용하는 알고리즘',name,'########')
    for score in ['accuracy','precision','recall','f1']:
        print(score)
        print('--')
        print(cross_val_score(model, train_input, train_target, scoring = score, cv=3))
```
