- merge

- 두 데이터프레임의 공통된 컬럼을 확인

```python
col_bank = bank.columns.to_list()  
col_bank_add = bank_add.columns.to_list() 
```

```python
# 두 리스트에 모두 존재하는 공통값을 반환하기
def isCommon(i):
  if i in col_bank:
    return i
```

```python
col_common = list(filter(isCommon, col_bank_add))
```

```python
print(col_common)
```

```python
pd.merge(bank, bank_add, on = col_common) # 두 데이터프레임에 공통된 컬럼의 값이 완전히 같은 데이터가 있는지 확인
```

```python
pd.merge(bank, bank_add, on = ['age', 'job', 'marital', 'default', 'housing']) # 두 데이터프레임에서 age, job, marital, default, housing 컬럼 값이 같은 테이터 병합
```

```python
bank_age = bank.set_index('age') # bank에서 age를 인덱스로 하는 새로운 데이터프레임 생성
```

```python
bank_age
```

```python
pd.merge(bank_age, bank_add, left_index = True, right_on = 'age') # age를 기준으로 병합하는데 
                                                                  # 왼쪽 데이터프레임에서는 age가 인덱스, 오른쪽 데이터프레임에서는 컬럼이므로 
                                                                  # 각각 left_index, right_on 매개변수를 이용
```

```python
pd.merge(bank, bank_add, on = 'age') # 위와 같은 결과
```

```python
pd.merge(bank, bank_add, on = 'age', how = 'inner') # age를 기준으로 병합하는데 양쪽 데이터프레임의 age 컬럼에서 공통된 값을  반영하여 병합
                                                    # 교집합 개념
```

```python
pd.merge(bank, bank_add, on = 'age', how = 'left') # age를 기준으로 병합하는데 왼쪽 데이터프레임의 age 컬럼 값을 모두 반영하여 병합
```

```python
pd.merge(bank, bank_add, on = 'age', how = 'right') # age를 기준으로 병합하는데 오른쪽 데이터프레임의 age 컬럼 값을 모두 반영하여 병합
```

```python
pd.merge(bank, bank_add, on = 'age', how = 'outer') # age를 기준으로 병합하는데 양쪽 데이터프레임의 age 컬럼 값을 모두 반영하여 병합
                                                    # 합집합 개념
```
