2. mpg 데이터에서 인덱스가 홀수인 행들만 추출

```python
mpg_odd = [] # 홀수를 저장할 빈 리스트 생성

for i in range(len(mpg)):
  if i%2 != 0: 
    mpg_odd.append(i)  # 홀수 인덱스를 추출하여 mpg_odd에 저장      

mpg.loc[mpg_odd, :]   # mpg_odd 변수를 loc에 적용하여 홀수 인덱스에 해당하는 행들만을 추출
```

3. gapminder 데이터에서 인덱스가 짝수인 행들만 추출

```python
gapminder_even = [] # 짝수를 저장할 빈 리스트 생성

for i in range(len(gapminder)):
  if i%2 == 0: 
    gapminder_even.append(i)  # 짝수 인덱스를 추출하여 gapminder_even에 저장      

gapminder.loc[gapminder_even, :]   # gapminder_even 변수를 loc에 적용하여 짝수 인덱스에 해당하는 행들만을 추출
```

4. 3번에서 생성한 데이터프레임을 새로운 변수에 저장하여 year, pop, gdpPercap의 평균 구하기

```python
even_gapminder = gapminder.loc[gapminder_even, :] # even_gapminder 에 3번에서 생성한 dataframe 
even_gapminder.loc[:, ['year', 'pop', 'gdpPercap']].mean() # loc 을 이용하여 year, pop, gdpPercap에 해당하는 컬럼을 추출하여 이에 대한 평균을 도출
```

5. 4번과 반대로 홀수 인덱스에 대해서 추출하여 4번의 결과와 비교

```python
gapminder_odd = [] # 홀수를 저장할 빈 리스트 생성

for i in range(len(gapminder)):
  if i%2 != 0: 
    gapminder_odd.append(i)  # 홀수 인덱스를 추출하여 gapminder_odd에 저장      

odd_gapminder = gapminder.loc[gapminder_odd, :]   # gapminder_odd 변수를 loc에 적용하여 홀수 인덱스에 해당하는 행들만을 추출하여 odd_gapminder 변수에 저장 
odd_gapminder.loc[:, ['year', 'pop', 'gdpPercap']].mean() # loc 을 이용하여 year, pop, gdpPercap에 해당하는 컬럼을 추출하여 이에 대한 평균을 도출
```

```python
even_mean = even_gapminder.loc[:, ['year', 'pop', 'gdpPercap']].mean()
odd_mean = odd_gapminder.loc[:, ['year', 'pop', 'gdpPercap']].mean()
even_mean < odd_mean 
```

8. mpg 데이터에서 2008년에 만들어진 자동차에서 복합연비가 좋은 4개를 추출

```python
mpg.assign(cby = lambda x : 0.55 * x.loc[:, 'cty'] + 0.45 * x.loc[:, 'hwy']).query('year == 2008').sort_values(by = 'cby', ascending = False).head(4) # lambda 함수와 loc을 이용하여 cby라는 복합연비 변수를 생성후 2008년에 해당하는 열에 대해 내림차순으로 정렬하여 위에서 4개를 추출
```
